```python
#以第一題為例
import math 
#輸入所需 x 值 , 所需誤差 TOL 值 , 最大迭代次數 M 值
x   = float(input('x 值 :'))
TOL = float(input('所需誤差 TOL 值 :'))
M   = int(input('最大迭代次數 M 值 :'))

#設定起始資料
x_0   = 1
i     = 1         # 累加從 i = 1 開始計算
y     = x-x_0
SIGN  = -1        # 起始正負符號
POWER = y         # 幕次累計器
TERM  = y         # 誤差值
SUM   = 0         # 累加從0開始計算
print('ln(1.5) = ','%.8f'%round(math.log(1.5),8))

#列印表格
print('================================================================')
print('n \t Pn \t\t TOL \t\t absolute error')
print('================================================================')

#迴圈運算
while i < M or i == M :
    SIGN = -SIGN
    SUM = SUM + SIGN*TERM
    POWER = POWER * y
    TERM = POWER/(i+1)
    print(i ,'\t' , '%.10f'%round(SUM,8) ,'\t' , '%.10f'%round(TERM,8),'\t' ,'%.10f'%round(abs(math.log(1.5)-SUM),8))  
    #math.log(x) i.e. ln(x)

    if abs(TERM) < TOL : #當下一項所產生之數值已小於所需誤差項
        print('================================================================')
        print('產生所需誤差之迭代次數 :',i,'次')
        break #強制跳離while迴圈
    i = i+1
```

```python
#以第一題為例-師版解答
import math 
#exact of ln(1.5)= 0.40546511
print('exact of ln(1.5) = %.8f'%round(math.log(1.5),8)) 

# INPUT  所需 x 值 , 所需誤差 TOL 值 , 最大迭代次數 M 值
x   = float(input('x   = '))  # x = 1.5
TOL = float(input('TOL = '))  # TOL = 0.0000001
M   = int(input('M   = '))    # M = 100

# Step 1 設定起始資料
x0 = 1
N = 1      # 累加從 i = 1 開始計算
y = x-x0
SUM = 0    # 累加從0開始計算
POWER = y  # 冪次累計器
TERM = y   # 誤差值
SIGN = -1  # 起始正負符號

# Step 2 迴圈運算
while N < M or N == M :
    
    # Step 3
    SIGN = - SIGN
    SUM = SUM +SIGN * TERM
    POWER = POWER * y
    TERM = POWER / (N+1)  # i=2 term
    
    # Step 4 
    if abs(TERM) < TOL:   #當下一項所產生之數值已小於所需誤差項
        print('The number of terms    for ln(1.5) is ', N)
        print('The approximated value for ln(1.5) is  %.8f'%round(SUM,8))
        print('The exact        value for ln(1.5) is  %.8f'%round(math.log(1.5),8))
        print('The absolute error     for ln(1.5) is  %.8f'%round(abs(math.log(1.5)-SUM),8))
        break             #強制跳離while迴圈
    
    # Step 5
    N = N + 1
    
# Step 6
print('The method is failed after ' ,M ,' iterations')
```

```python
#以第二題為例-師版解答
import math 
#exact of pi= 3.14159265
print('exact of pi = %.8f'%round(math.pi,8)) 

# INPUT  所需 x 值 , 所需誤差 TOL 值 , 最大迭代次數 M 值
#x   = float(input('x   = '))  # x = 1
#TOL = float(input('TOL = '))  # TOL = 0.0000001
#M   = int(input('M   = '))    # M = 100000000

x=1
TOL = 0.0000001
M = 100000000

# Step 1 設定起始資料
x0 = 0
N = 1      # 累加從 i = 1 開始計算
y = x-x0
SUM = 0    # 累加從0開始計算
POWER = 1/y  # 冪次累計器
TERM = 0   # 誤差值
SIGN = 1  # 起始正負符號

# Step 2 迴圈運算
while N < M or N == M :
    
    # Step 3
    SIGN = - SIGN
    SUM = SUM +SIGN * TERM
    POWER = POWER * y
    TERM = POWER / (2*N-1)  # i=2 term
    
    # Step 4 
    if abs(math.pi - 4*SUM) < TOL: #當下一項所產生之數值已小於所需誤差項
        print('The number of terms    for pi is ', N)
        print('The approximated value for pi is  %.8f'%round(4*SUM,9))
        print('The exact        value for pi is  %.8f'%round(math.pi,9))
        print('The absolute error     for pi is  %.8f'%round(abs(math.pi-4*SUM),9))
        break #強制跳離while迴圈
    
    # Step 5
    N = N + 1
    
# Step 6
print('The method is failed after ' ,M ,' iterations')
```
